```python
#Traversing
arr = [3,19,2,99,5,4,55,19,120]
for i in arr:
    print(i)
```

```python
# finding max value
max_val = arr[0]
for num in range(1,len(arr)):
    if arr[num] > max_val:
        max_val = arr[num]
    print("num:" , arr[num] ,"max:" , max_val)

print(max_val)        
```

```python
# finding min value
min_val = arr[0]
for i in range(1,len(arr)):
    if arr[i] < min_val:
        min_val = arr[i]
    print("n:",arr[i],"min:",min_val)
min_val
```

```python
#Counting even no 
count = 0
for i in range(0,len(arr)):
    if arr[i]%2==0:
        count+=1
    print("Num: ",arr[i],"Count: ",count)
print(count)    
```

```python
# checking if it is sorted
arr = [1,2,3,4,5,6]
flag = True
for i in range(1,len(arr)):
    if arr[i] < arr[i-1]:
        flag = False
        break

    print("Num: ",arr[i-1],arr[i], flag)
print(flag)
```
